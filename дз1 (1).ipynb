# 2.1 Задача (sleep equation)

# 2.1.1 Пример 1

```python
import numpy as np # linear algebra 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
# stats models: regression fitting
import statsmodels.formula.api as smf
# data visualization
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
sns.scatterplot(data=df, y='sleep', x='totwrk')
```

```python
sns.regplot(data=df, y='sleep', x='totwrk', ci=None, line_kws={"color": "r"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='sleep~totwrk', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

###### Параметры оптимальной прямой sleep на totwrk без константы ######

```python
# подгонка прямой без константы
fitted_line = smf.ols(formula='sleep~ - 1 + totwrk', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

# 2.2 Задача (labour equation)

# 2.2.1 Пример 1

```python
import numpy as np # linear algebra 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
# stats models: regression fitting
import statsmodels.formula.api as smf
# data visualization
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df
```

```python
sns.scatterplot(data=df, x='capital', y='output')
```

```python
sns.regplot(data=df, x='capital', y='output', ci=None, line_kws={"color": "red"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='output~capital', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

#### Параметры оптимальной прямой output на capital без константы

```python
# подгонка прямой без константы
fitted_line = smf.ols(formula='output~-1 + capital', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

#### оптимальная прямая log(output) на log(capital)

```python
sns.regplot(x=np.log(df['capital']), y=np.log(df['output']), ci=None, line_kws={"color": "red"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='np.log(output)~np.log(capital)', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

#### оптимальная прямая log(output) на log(capital) без константы

```python
# подгонка прямой без константы
fitted_line = smf.ols(formula='np.log(output)~-1+np.log(capital)', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

# 2.2.1 Пример 2

#### Оптимальная прямая output на labour

#### Оптимальная прямая output на labour без константы

#### Оптимальная прямая log(output) на log(labour)

```python
sns.regplot(x=np.log(df['labour']), y=np.log(df['output']), ci=None, line_kws={"color": "black"})
```

```python
# подгонка прямой без константы
fitted_line = smf.ols(formula='np.log(output)~np.log(labour)', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```
